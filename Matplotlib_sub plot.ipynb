```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,9])
plt.plot(x,y)
plt.show()
```

```python
#double plot
x = np.array([1,2,3,4])
y = np.array([1,2,3,9])
plt.plot(x,y)
plt.show()

x = np.array([1,2,3,4])
y = np.array([1,2,3,4])
plt.plot(x,y, color="red")
plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,9])
plt.plot(x,y,color = "b")
plt.subplot(2,2,1)

a= np.array([1,2,3,4])
b = np.array([1,2,3,4])
plt.plot(a,b, color="red")
plt.subplot(2,2,2)

c= np.array([6,2,3,4])
d = np.array([1,22,3,40])
plt.bar(c,d, color="green")
plt.subplot(2,2,3)

e= np.array([60,20,10,5])
f = ["apple","mango","banana","graph"]
plt.pie(e)
plt.subplot(2,2,4)

plt.show()
```
