```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]
z = pd.DataFrame({"x":x, "y":y})
sns.lineplot(x="x" , y= "y", data=z)

plt.show()
```
