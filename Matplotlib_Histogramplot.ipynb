```python
#cummulative parameter
no = [58, 49, 48, 21, 57, 24 ,53, 53, 51, 67, 55, 32, 21, 20, 55, 60, 24, 56, 24, 53, 56, 31, 31, 48,
 69 ,52, 20 ,22 ,52, 20 ,46, 45, 56, 64, 50, 48, 24 ,23, 48, 63, 68, 29, 20, 47, 49, 61, 26, 24,
 22, 49]
b = [10,20,30,40,50,60,70]
plt.hist(no, bins=b ,color="red",edgecolor="black", cumulative=-1 )
plt.title("Histogram")
plt.xlabel("Year")
plt.ylabel("sales")
plt.show()
```
