```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([10,30,25,15])
y = ["java","python","C","html"]
plt.pie(x)
plt.show()
```

```python
#adding labels
x = np.array([10,30,25,15])
y = ["java","python","C","html"]
plt.pie(x, labels=y)
plt.show()
```

```python
#more parameters
x = np.array([10,30,25,15])
y = ["java","python","C","html"]
e = [0.0,0.1,0.0,0.0]
c = ["r","g", "b","m"]
plt.pie(x, labels=y , explode = e , colors=c , autopct="%0.2f%%", shadow=True)
plt.legend(title="most popular language")
plt.show()
```
