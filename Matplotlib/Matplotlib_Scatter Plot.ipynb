```python
#scatter plot
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([5,7,8,35,])
y = np.array([99,109,65,78])
plt.scatter(x,y)
plt.show()
```

```python
x = np.array([5,7,8,35,3,9,56,7,56,2,3,5])
y = np.array([99,109,65,78,99,100,56,78,102,105,110,140])
plt.title("Scatter Plot")
plt.xlabel("no of cars")
plt.ylabel("speed of cars")
plt.scatter(x,y,color="red")
plt.show()
```

```python
import numpy as np
import  matplotlib.pyplot as plt
x = np.array([5,7,8,35,3])
y = np.array([99,109,65,78,140])
colors = np.array(["green","black","red","blue","pink"])
sizes = np.array([50,100,30,60,80])
plt.title("Scatter Plot")
plt.xlabel("no of cars")
plt.ylabel("speed of cars")
plt.scatter(x,y, c=colors , s=sizes, marker="*")
plt.show()
```

```python
#import numpy as np
#import  matplotlib.pyplot as plt
x = np.array([5,7,8,35,3])
y = np.array([99,109,65,78,140])
colors = np.array(["green","black","red","blue","pink"])
sizes = np.array([50,100,30,60,80])
plt.title("Scatter Plot")
plt.xlabel("no of cars")
plt.ylabel("speed of cars")
plt.scatter(x,y, c=colors , s=sizes, marker="*", edgecolor="b", linewidth= 1)
plt.show()
```

```python
x = np.array([5,7,8,35,3,10,19])
y = np.array([99,109,65,78,140,55,86])
colors = np.array([23,78,90,55,77,10,98])
sizes = np.array([50,100,90,60,80,77,99])
plt.title("Scatter Plot")
plt.xlabel("no of cars")
plt.ylabel("speed of cars")
#plt.colorbar()
plt.scatter(x,y, c=colors , cmap='viridis', s=sizes, edgecolor="b", linewidth= 1)
plt.show()
```
