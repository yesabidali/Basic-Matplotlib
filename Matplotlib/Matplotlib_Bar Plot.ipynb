```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["Python","C","Rush","Java"]
y = [85,45,70,35]
c = ["g","b","y","m"]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("Language Popularity graph")
plt.bar(x,y,width=0.6,color=c,align="center",edgecolor="b",label="popularity")
plt.legend()
plt.show()
```

```python
#overlaping bar graph
x = ["Python","C","Rush","Java"]
y = [85,45,70,35]
z = [20,80,3,75]
c = ["g","b","y","m"]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("Language Popularity graph")
plt.bar(x,y,width=0.6,color=c,align="center",edgecolor="b",label="popularity")
plt.bar(x,z,width=0.6,align="center",edgecolor="r",label="popularity before 2010")
plt.legend()
plt.show()
```
